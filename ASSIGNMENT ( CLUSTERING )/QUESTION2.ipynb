```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.cluster import AgglomerativeClustering
import sklearn
from sklearn.preprocessing import RobustScaler,StandardScaler,MinMaxScaler
```

```python
data=pd.read_csv("D:/DATA SCIENCE/ASSIGNMENT/QUESTION/CLUSTERING/EastWestAirlines.csv")
```

```python
scaler=RobustScaler()
rb_data=scaler.fit_transform(data)
```

```python
rb_data
```

```python
no=range(1,20)
k=[]

for i in no:
    model=KMeans(n_clusters=i)
    model.fit(rb_data)
    k.append(model.inertia_)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(no,k,cmap="black",s=50)
plt.plot(no,k,linestyle="-")
plt.title("elbow technique" )
plt.xticks(no)
plt.yticks(k)
```

```python
model=KMeans(n_clusters=3)
model.fit_predict(rb_data)
```

```python
model.cluster_centers_
```

```python
cl=model.fit_predict(rb_data)
```

```python
data['cluster']=cl
```

```python
data
```

```python
data.groupby(['cluster'])["ID#"].apply(lambda x:tuple(x))
```

```python
from sklearn.metrics import silhouette_score
```

```python
stnd_scalar=MinMaxScaler()
data=stnd_scalar.fit_transform(data)
```

```python
data
```

```python
n=range(2,11)
l=[]

for i in n:
    model=AgglomerativeClustering(n_clusters=i,affinity="l2",linkage="single")
    l.append(silhouette_score(data,model.fit_predict(data)))
```

```python
plt.figure(figsize=(10,6))
plt.scatter(n,l,color="red")
plt.plot(n,l,linestyle="-")
```

```python
model=AgglomerativeClustering(n_clusters=4,affinity="l2",linkage="complete")
a=model.fit_predict(data)
plt.hist(a)
```

```python
model=AgglomerativeClustering(n_clusters=6,affinity='l2',linkage="complete")
b=model.fit_predict(data)
plt.hist(b)
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
from scipy.spatial.distance import pdist
```

```python
x=pdist(data,metric='euclidean')
l=linkage(x,'ward')

plt.figure(figsize=(20,8))
dendrogram(l,orientation="top",distance_sort=True,show_leaf_counts=True)
```

```python
from sklearn.datasets import make_moons
```

```python
x,y=make_moons(100)
```

```python
plt.figure(figsize=(10,4))
plt.style.use("seaborn-darkgrid")
plt.scatter(x[:,0],x[:,1])
```

```python
model=DBSCAN(min_samples=2,eps=0.2)
model.fit_predict(x)

plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=model.fit_predict(x),cmap=plt.cm.rainbow_r)
```

```python
model=KMeans(n_clusters=3)
model.fit_predict(x)

plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=model.fit_predict(x),cmap=plt.cm.rainbow_r)
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1])
```

```python
model=AgglomerativeClustering(n_clusters=6,affinity="l2",linkage="average")
model.fit_predict(x)

plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=model.fit_predict(x),cmap=plt.cm.rainbow_r)
```

```python
model=AgglomerativeClustering(n_clusters=2,affinity="l2",linkage="average")
model.fit_predict(x)

plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=model.fit_predict(x),cmap=plt.cm.rainbow_r)
```
