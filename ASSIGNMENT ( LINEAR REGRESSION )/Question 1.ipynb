```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.linear_model import LinearRegression
import statsmodels.formula.api as smf
```

```python
data=pd.read_csv("D:/DATA SCIENCE/ASSIGNMENT/QUESTION/LINEAR REGESSION/delivery_time.csv")
```

```python
data
```

```python
plt.figure(figsize=(10,4))
plt.scatter(data=data,x='Delivery Time',y='Sorting Time')
```

```python
data.corr()
```

```python
d_time=data['Delivery Time']
s_time=data['Sorting Time']
```

```python
model=smf.ols("d_time~s_time",data=data).fit()
```

```python
model.summary()
```

```python
plt.figure(figsize=(10,4))
plt.style.use("seaborn-darkgrid")

plt.scatter(data['Sorting Time'],data['Delivery Time'],color="black")
#plt.plot(model.predict(data['Delivery Time']))
plt.plot(data['Sorting Time'],model.fittedvalues,label='Fitted value',color='red')
plt.legend()

plt.xlabel('sorting time')
plt.ylabel('Delivery time')
```

```python
model1=smf.ols("d_time~np.log(s_time)",data=data).fit()
```

```python
model1.summary()
```

```python
model2=smf.ols("np.log(d_time)~(s_time)",data=data).fit()
```
