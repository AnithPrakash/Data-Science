```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.formula.api as smf

```

```python
data=pd.read_csv("D:/DATA SCIENCE/ASSIGNMENT/QUESTION/LINEAR REGESSION/Salary_Data.csv")
```

```python
data
```

```python
plt.figure(figsize=(10,6))
plt.style.use("seaborn-darkgrid")
plt.scatter(x='YearsExperience',y='Salary',data=data,color="blue")
plt.xlabel("YearsExperience")
plt.ylabel("salary")
```

```python
data.corr()
```

```python
model=smf.ols("YearsExperience~Salary",data=data).fit()
```

```python
model.summary()
```

```python
plt.figure(figsize=(10,5))

plt.scatter(x='YearsExperience',y='Salary',data=data,color="blue")

plt.plot(model.predict(data['Salary']),data['Salary'])

plt.xlabel("YearsExperience")

plt.ylabel("salary")

plt.title("Salary vs Year Experience")
```

```python
plt.scatter(data['Salary'],model.predict(data['Salary']),s=5)
```
