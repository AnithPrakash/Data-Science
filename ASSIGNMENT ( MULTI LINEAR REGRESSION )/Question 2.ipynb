```python
import statsmodels.formula.api as smf
```

```python
model=smf.ols("Profit~R_D_S+Admin+market_spend+State",data=data).fit()
model.rsquared
```

```python
radmin=smf.ols('Admin~R_D_S+market_spend+State',data=data).fit().rsquared
vadmin=1/(1-radmin)
```

```python
rstate=smf.ols('State~Admin+R_D_S+market_spend',data=data).fit().rsquared
vstate=1/(1-rstate)
```

```python
rms=smf.ols('market_spend~Admin+R_D_S+State',data=data).fit().rsquared
vms=1/(1-rms)
```

```python
rrd=smf.ols('R_D_S~Admin+market_spend+State',data=data).fit().rsquared
vrd=1/(1-rrd)
```

```python
d1={"value":['admin','state','market spend','r_d_s'],'VIF':[vadmin,vstate,vms,vrd]}
vif=pd.DataFrame(d1)
vif
```

```python
print(smf.ols("Admin~State+market_spend+R_D_S",data=data).fit().rsquared)
print(smf.ols("State~Admin+R_D_S+market_spend",data=data).fit().rsquared)
print(smf.ols("market_spend~Admin+State+R_D_S",data=data).fit().rsquared)
print(smf.ols("R_D_S~Admin+State+market_spend",data=data).fit().rsquared)
```

```python
model=smf.ols('R_D_S~market_spend+np.log(Admin)+State+pow(Admin,2)',data=data).fit()
model.summary()
```

```python
plt.hist(model.resid,bins=10)
```

```python
plt.boxplot(model.resid)
```

```python
import statsmodels.api as sm
```

```python
qplot=sm.qqplot(model.resid,line="q")
plt.title("qq plot")
plt.show()
```

```python
def g(v):
  return (v - v.mean())/v.std()
```

```python
plt.scatter(g(model.fittedvalues),g(model.resid))
plt.title("residual plot")
#plt.xlabel("")
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```

```python
plt.scatter(data['Admin'],g(model.resid))
plt.show()
```
