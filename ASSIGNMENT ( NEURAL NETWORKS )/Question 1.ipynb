```python
import tensorflow as tf
import numpy as np 
import pandas as pd
import sklearn
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing

```

```python
data=pd.read_csv("D:/DATA SCIENCE/ASSIGNMENT/QUESTION/NEURAL NETWORK/forestfires.csv")
```

```python
data
```

```python
data['size_category']=preprocessing.LabelEncoder().fit_transform(data['size_category'])
```

```python
data['month']=preprocessing.LabelEncoder().fit_transform(data['month'])
```

```python
data['day']=preprocessing.LabelEncoder().fit_transform(data['day'])
```

```python
data
```

```python
#data=data.drop(columns=['month','day'])
```

```python
data
```

```python
x = data.iloc[:,:30]
y = data.iloc[:,30]
```

```python
x
```

```python
y
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(44,input_dim=30,activation='relu'))
model.add(tf.keras.layers.Dense(26,activation='sigmoid'))
model.add(tf.keras.layers.Dense(1,activation='tanh'))
```

```python
model.summary()
```

```python
model.compile(loss ='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

```python
history=model.fit(x, y, validation_split=0.20, epochs=50, batch_size=100)
```

```python
score=model.evaluate(x,y)
print('%s: %.2f%%' % (model.metrics_names[1], score[1]*100))
```
